# Generating Question and Answer

```python
'''
Question:
What is the timing for the '{some_event_name}'?
Where is '{some_event_name}' taking place?
Who can attend '{some_event_name}'?
Can you tell me more about the '{some_event_name}'?

Answer:

Timings for '{some_event_name}' are '{some_time}'
'{some_location}'
'{some_event_name}'
'{some_details_summary}'

'''
```

Generating 4 questions per entry

```python
combined_df.shape
```

```python
qna_list = []
# Iterate over each row in Courses_df
for index, row in combined_df.iterrows():
    
    # Generate 4 questions and answers for each row
    question = f'What is the timing for the {row["EventName"]}?'
    answer = f'Timings for {row["EventName"]} are {row["Time"]} on {row["Date"]}.'
    docid=f'{row["DocID"]}'
    qna_list.append({"Question": question, "Ref_Answer": answer, "DocID": docid})
    
    question = f'Who can attend {row["EventName"]}?'
    answer = f'{row["Participants"]}.'
    docid=f'{row["DocID"]}'
    qna_list.append({"Question": question, "Ref_Answer": answer, "DocID": docid})
    
    question = f'Where is {row["EventName"]} taking place?'
    answer = f'{row["Location"]}'
    docid=f'{row["DocID"]}'
    qna_list.append({"Question": question, "Ref_Answer": answer, "DocID": docid})
    
    question = f'Can you tell me more about the {row["EventName"]}?'
    answer = f'{row["Details"]}'
    docid=f'{row["DocID"]}'
    qna_list.append({"Question": question, "Ref_Answer": answer, "DocID": docid})
    
    
print(len(qna_list))
QnA_df = pd.DataFrame(qna_list)
QnA_df['Category'] = 'Events'
# Reset index of QnA_df
QnA_df.reset_index(drop=True, inplace=True)
```

```python
# Define filenames for questions and answers files
questions_filename = "questions.txt"
answers_filename = "answers.txt"

# Open files in append mode
with open(questions_filename, 'a') as questions_file, open(answers_filename, 'a') as answers_file:
    # Iterate over each row in the DataFrame
    for index, row in QnA_df.iterrows():
        question = row['Question']
        answer = row['Ref_Answer']

        # Write question and answer to their respective files
        questions_file.write(question + "\n")
        answers_file.write(answer + "\n")
```
